```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from yellowbrick.text import TSNEVisualizer
```

## Espanha

```python
url_esp = 'https://raw.githubusercontent.com/aaasatoou/bootcamp_blue_Projeto2/main/Estudo%20COVID%20-%20Insights%20e%20Previs%C3%B5es/dados/Silver/datasets/Titulo%20Noticias%20Espanha%20Corrigido.csv'
not_esp = pd.read_csv(url_esp)
not_esp.drop('Unnamed: 0',axis=1,inplace=True)
not_esp.head()
```

```python
not_esp['Data'] = pd.to_datetime(not_esp['Data'],format='%Y-%m-%d')
not_esp['Ano'] = not_esp['Data'].dt.year
not_esp.head()
```

```python
sns.countplot(x='Ano',data=not_esp)
```

```python
plt.figure(figsize=(12,6),dpi=150)
plt.title('Quantidade de Noticias por Cluster nos Anos 2020 e 2021 na Espanha',
          fontsize=14)
sns.countplot(x='Cluster',data=not_esp,hue='Ano')
plt.ylabel('Quantidade');
```

```python
count_vec = CountVectorizer()
mat_esp = count_vec.fit_transform(not_esp.Titulo)

tsne = TSNEVisualizer(size=(1080, 720),title='Projeção TNSE dos Clusters da Espanha')
tsne.fit(mat_esp, not_esp.Cluster)
tsne.show()
```
